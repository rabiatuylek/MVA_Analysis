```python
#Multiple Regression
# Data from https://archive.ics.uci.edu/ml/datasets/combined+cycle+power+plant
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

dataset = pd.read_csv('/Users/Rabia/OneDrive/Masaüstü/Data_Science/Multiple_Regression/MRData.csv')
```

```python
dataset.head()
#dataset.head(a) a değeri kaç satır varsa o kadar satır data bilgisi verir.
```

```python
#define x and y
# 2 different dataset is created. >>> x & y
x = dataset.drop(['PE'],axis=1).values  # to cancel vertical that We want
y = dataset['PE'].values # to show vertical that We want
```

```python
print(x)  
```

```python
print(y)
```

```python
#Split the dataset in training set and test set
# %30 test, %70 train

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 0)
```

```python
#Train the model on the Training test
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
#Print the test set results
y_prediction = lr.predict(x_test)
print(y_prediction)
```

```python
print(x[0])
```

```python
lr.predict([x[0]]) # predict PE for first satır
#PE was 463.27 but now predicted value is 467.348
```

```python
lr.predict([x[234]]) # predict PE for 234. satır
```

```python
#Evaluate Model
from sklearn.metrics import r2_score
r2_score(y_test, y_prediction)
```

```python
#Plot the results
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
plt.scatter(y_test,y_prediction)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted with Multiple Regression')
```
